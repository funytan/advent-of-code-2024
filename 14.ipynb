```python
num_x = 101
num_y = 103
steps = 100
with open('inputs/input_14', 'r') as f:
    lines = f.readlines()
    final_pos = []
    for line in lines:
        p_x, p_y = list(map(int, line.split('=')[1][:-2].split(',')))
        v_x, v_y = list(map(int, line.split('=')[2][:-1].split(',')))
        for step in range(steps):
            p_x, p_y = p_x+v_x, p_y+v_y
            p_x %= num_x
            p_y %= num_y
        final_pos.append((p_x, p_y))

quadrants = {
    1: (0, num_x//2-1, 0, num_y//2-1),
    2: (num_x//2+1, num_x-1, 0, num_y//2-1),
    3: (0, num_x//2-1, num_y//2+1, num_y-1),
    4: (num_x//2+1, num_x-1, num_y//2+1, num_y-1),
}

res = defaultdict(int)
for x, y in final_pos:
    for quad, bounds in quadrants.items():
        if bounds[0]<=x<=bounds[1] and bounds[2]<=y<=bounds[3]:
            res[quad]+=1

print(res[1]*res[2]*res[3]*res[4])
    
    
```

```python
def draw(pos, num_x, num_y):
    pos = set(pos)
    for y in range(num_y):
        row = []
        for x in range(num_x):
            if (x,y) in pos:
                row.append('*')
            else:
                row.append('.')
        print(''.join(row))

# for this test, we assume the heuristic that the tree is more or less symmetrical around the center.
# We try to test this by taking the difference between the mass of robots that share a similar y coordinate but find themselves on oppostie sides of the center vertical
def test(pos, num_x, num_y):
    center = sum([p[0] for p in pos])/len(pos)
    difference_in_moment = 0
    y_pos = defaultdict(set)
    for x, y in pos:
        y_pos[y].add(x) # to account for overlaps
    for y in range(num_y):
        # if y == 1:
        #     print(len(y_pos[y]))
        # if y in y_pos and len(y_pos[y])%2==0:
        left = [x for x in y_pos[y] if x<center]
        right = [x for x in y_pos[y] if x>center]
        num_left = len(left)
        num_right = len(right)
        if len(left) == len(right):
            left_mom = sum([center-x for x in left])
            right_mom = sum([x-center for x in left])
            difference_in_moment += abs(left_mom- right_mom)
    return difference_in_moment 
                
        
    
    
```

```python
def run(steps):
    num_x = 101
    num_y = 103
    steps = steps
    with open('inputs/input_14', 'r') as f:
        lines = f.readlines()
        pos = []
        vel = []
        for line in lines:
            p_x, p_y = list(map(int, line.split('=')[1][:-2].split(',')))
            v_x, v_y = list(map(int, line.split('=')[2][:-1].split(',')))
            pos.append((p_x, p_y))
            vel.append((v_x, v_y))
        for i in range(1, steps+1):
            new_pos = []
            for p, v in zip(pos, vel):
                p_x, p_y = p[0]+v[0], p[1]+v[1]
                p_x %= num_x
                p_y %= num_y
                new_pos.append((p_x, p_y))
            pos=new_pos
            diff_in_moment = test(pos, num_x, num_y)
            if diff_in_moment < 500:
                print(i, diff_in_moment)
            # if i == 4038:
                print(f"Step: {i}")        
                draw(pos, num_x, num_y)
                clear_output(wait=True)
                time.sleep(1)
                # break
            assert len(pos)==500
```
